When the confidence levels of first,second and third prediction change, the frequency of dogs in the dataset changes accordingly

This will be demonstrated in the next line of codes

filtering data by rows where the algorithm's prediction is equal to or greater than 95% and the prediction is a 
breed of dog and the tweet wasn't retweeted

```python
sb.countplot(y = 'p1', data = filtered_df,  color = base_color, order = gen_order[:10]);
plt.xticks(rotation = 90);
```

In the next line of code I will be reducing the confidence levels to equal or greater than 90% to identify visible difference

```python
sb.countplot(y = 'p1', data = filtered_df_1,  color = base_color, order = gen_order[:10]);
plt.xticks(rotation = 90);
```

The above chart shows that when the confidence level is written as a code to be greater than or equal to 90%, the order of frequency of dogs changed, the new chart now shows Golden Retriever, Pug, Pembroke, Samoyed, Labrador Retriever, Pomeranian, Chihuahua, Fench Bulldog, Chow and Blenheim Spaniel in that order of frequency,

`Pembroke`: Pembroke was third in the order of frequency in the first chart, and this was still the case in the second chart

`Golden Retriever`: This was the fourth dog in the order of frequency in the first chart and it became the first dog in the order of frequency when the confidence level changes as displaed in the second chart
